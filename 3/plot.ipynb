```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.rcParams['figure.dpi'] = 442
```

```python
def read_file(file):
    iou = []
    with open("res/" + file) as f:
        for line in f:
            iou.append(float(line.split(" ")[1][:-1]))
    return iou
```

```python
epochs = list(range(2,82,2))
ious = []

plt.title("Intersection Over Union (IoU) vs epochs")
plt.xlabel("Epoch")
plt.ylabel("IoU")
plt.grid(True)
plt.xlim([2, 80])
for c in range(4):
    ious.append(read_file("testIoU" + str(c + 1) + ".txt"))
    plt.plot(epochs, ious[c], label="Configuration {}".format(c + 1))
    print("Maximum IoU value for configuration {}: {}".format(c + 1, max(ious[c])))
plt.legend()
plt.savefig("iou.png")
plt.show()
```
