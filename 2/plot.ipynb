```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import json
import os
```

```python
mpl.rcParams['figure.dpi'] = 442
```

```python
def plot(param, title):
    path = "results.%s/" % param
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Validation Accuracy")
    plt.grid(True)
    for filename in os.listdir(path):
        with open(path + filename) as f:
            data = json.load(f)
            learning_curve = data["learning_curve"]
            plt.plot([p["epoch"] for p in learning_curve],
                     [p["val_categorical_accuracy"] for p in learning_curve],
                     label='{}={}'.format(param, data[param]))
            plt.legend()
    plt.savefig(param)
    plt.show()
```

# Learning Rate

```python
plot('lr', 'Learning Curves for different learning rates')
```

# Filter Size

```python
plot('filter_size', 'Learning Curves for different filter sizes')
```
